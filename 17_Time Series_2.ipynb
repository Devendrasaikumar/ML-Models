#### <font color='magenta'>**SARIMA** </font>

**S**easonal **A**uto**R**egressive **I**ntegrated **M**oving **A**verage

- The SARIMA (Seasonal AutoRegressive Integrated Moving Average) model is an extension of the ARIMA model that explicitly accounts for seasonality in time series data. 

- SARIMA is used when a time series shows seasonal effectspatterns that repeat at regular intervals, like monthly sales peaks during the holiday season or daily temperature variations due to the time of year.

 The key difference between ARIMA and SARIMA is the inclusion of seasonal terms. In addition to the non-seasonal components of ARIMA (AutoRegressive - AR, Integrated - I, and Moving Average - MA), SARIMA also includes seasonal components that account for periodic fluctuations.

#### <font color='magenta'> </font>

The SARIMA model is denoted as SARIMA **(p, d, q)(P, D, Q)[s]**. Here's what each parameter represents:

**(p, d, q)**: These are the non-seasonal parameters, similar to the ARIMA model:

**p**: The number of lag observations included in the model (AutoRegressive component).

**d**: The number of times the data is differenced to make it stationary (Integrated component).

**q**: The number of lagged forecast errors in the prediction equation (Moving Average component).

**(P, D, Q)**: These are the seasonal counterparts to the non-seasonal parameters:

**P**: The number of seasonal lags in the AutoRegressive component.

**D**: The number of seasonal differences required to make the series stationary.


**Q**: The number of seasonal lags in the Moving Average component.

**s**: The length of the seasonal cycle (e.g., for monthly data with yearly seasonality, s = 12).

#### <font color='magenta'>**Prophet** </font>

- Prophet is an open-source forecasting tool developed by Facebook that is designed to handle time series data with strong seasonal patterns and missing data

- Unlike traditional models like ARIMA or SARIMA, Prophet is easy to use, flexible, and robust to missing data and outliers.

**Key Features of Prophet** :-

**Handles missing data**: Prophet can handle missing data points or data irregularities (such as holidays or special events) without needing imputation or data filling.

**Seasonality**: It can model multiple types of seasonality (e.g., yearly, weekly, or daily).

**Holiday Effects**: Prophet can incorporate the effect of holidays or special events that might affect the time series.

**Flexibility**: It allows flexibility in the trend modeling (such as logistic growth or linear trends) to capture various patterns of growth.

#### <font color='magenta'>**ETS** </font>

- Exponential Smoothing State Space Model

- It is based on the idea of exponentially smoothing historical data to forecast future values, where more recent observations are given more weight than older ones. 

 The ETS model decomposes the time series into three main components:

**Error (E)**: The random fluctuation or noise in the data.

**Trend (T)**: The long-term increase or decrease in the series.

**Seasonality (S)**: The periodic, repeating patterns in the series.

ETS is flexible in capturing both additive and multiplicative trends and seasonality, which allows it to model both constant and proportionally growing effects in the data.

#### <font color='magenta'>  </font>

- **Error (E)**: This captures the random fluctuations in the data.

**Additive error** means that the error term is constant (it does not change over time).

**Multiplicative error** means that the error grows in proportion to the level of the series.

#### <font color='magenta'>  </font>

- **Trend (T)**: The trend models the long-term direction of the series, either increasing or decreasing over time.

**Additive trend** means that the trend increases or decreases by a constant amount over time.

**Multiplicative trend** means that the trend increases or decreases by a constant percentage over time.

#### <font color='magenta'>  </font>

- **Seasonality (S)**: The seasonal component models the repeating patterns that occur at fixed intervals (e.g., yearly, monthly, weekly).

**Additive seasonality** means that the seasonal fluctuation is constant over time.

**Multiplicative seasonality** means that the seasonal fluctuations change in proportion to the level of the series.

#### <font color='magenta'>**LSTM** </font>

- **L**ong **S**hort-**T**erm **M**emory

Designed to handle and predict sequences of data, such as time series data, text, or speech. LSTM is a type of deep learning model that is particularly effective for handling long-term dependencies in sequences, meaning it can remember information from earlier parts of a sequence (e.g., time series data) and use it to make predictions about future points in the sequence.

*The architecture of an LSTM consists of a chain of repeating units, each consisting of three gates*:

**Forget Gate**: Decides what information should be discarded from the cell state.

**Input Gate**: Updates the cell state with new information.

**Output Gate**: Determines what the next hidden state will be.

The LSTM unit has cell states that can pass information through time without it being forgotten, making it capable of learning long-term dependencies in the data.

## <font color='red'>**Summary of these models** </font>

**In Summary**:

- **ARIMA and SARIMA** are great for modeling time series data with clear auto-correlation, trends, and seasonality (especially if the data is stationary or made stationary).
- **Prophet** is ideal for handling multiple seasonalities and irregular events (like holidays) in data, making it suitable for business applications.
- **ETS** works well for stable data with clear trend and seasonality, providing a simple and fast approach.
- **LSTM** is best for complex, nonlinear patterns in data, especially when you have a large amount of sequential data and need to capture long-term dependencies.

#### <font color='magenta'>1. Import Libraries and Generate Data </font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet 
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler

# Generate sample data (monthly sales data for 3 years)
np.random.seed(0)
months = pd.date_range(start='2019-01-01', periods=36, freq='M')
sales = np.random.randint(100, 200, size=(36,)) + (np.linspace(0, 50, 36))  # trend + noise

# Create DataFrame
data = pd.DataFrame({'Date': months, 'Sales': sales})
data.set_index('Date', inplace=True)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Sales'], label="Actual Sales")
plt.title('Monthly Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

 #### <font color='magenta'>2. ARIMA Model </font>

```python
# Split data into training and test sets (80% training, 20% test)
train = data[:30]
test = data[30:]

# Fit ARIMA model (order (p,d,q))
model_arima = ARIMA(train, order=(1,1,1))
model_arima_fit = model_arima.fit()

# Make predictions
forecast_arima = model_arima_fit.forecast(steps=6)
forecast_arima_index = pd.date_range(start=test.index[0], periods=6, freq='M')

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(train.index, train['Sales'], label="Training Data")
plt.plot(test.index, test['Sales'], label="Test Data")
plt.plot(forecast_arima_index, forecast_arima, label="ARIMA Forecast", linestyle='--')
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Evaluate RMSE
rmse_arima = np.sqrt(mean_squared_error(test['Sales'], forecast_arima))
print(f"ARIMA RMSE: {rmse_arima:.2f}")
```

#### <font color='magenta'>3. SARIMA Model</font>

```python
# Fit SARIMA model (order (p,d,q)(P,D,Q,s))
model_sarima = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12))
model_sarima_fit = model_sarima.fit()

# Make predictions
forecast_sarima = model_sarima_fit.forecast(steps=6)
forecast_sarima_index = pd.date_range(start=test.index[0], periods=6, freq='M')

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(train.index, train['Sales'], label="Training Data")
plt.plot(test.index, test['Sales'], label="Test Data")
plt.plot(forecast_sarima_index, forecast_sarima, label="SARIMA Forecast", linestyle='--')
plt.title('SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Evaluate RMSE
rmse_sarima = np.sqrt(mean_squared_error(test['Sales'], forecast_sarima))
print(f"SARIMA RMSE: {rmse_sarima:.2f}")
```

#### <font color='magenta'>4. Prophet Model  </font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from prophet import Prophet

# Prepare data for Prophet
prophet_data = data.reset_index()  # Reset index, assuming 'data' has 'Date' and 'Sales' columns
prophet_data.columns = ['ds', 'y']  # Prophet expects 'ds' for dates and 'y' for target variable

# Fit Prophet model
model_prophet = Prophet(yearly_seasonality=True)
model_prophet.fit(prophet_data)  # Fit model on the entire data

# Create future dataframe (forecast for the next 6 months)
future = model_prophet.make_future_dataframe(periods=6, freq='M', include_history=False)

# Make predictions
forecast_prophet = model_prophet.predict(future)

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(prophet_data['ds'], prophet_data['y'], label="Actual Sales")
plt.plot(forecast_prophet['ds'], forecast_prophet['yhat'], label="Prophet Forecast", linestyle='--')
plt.title('Prophet Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Evaluate RMSE (ensure 'test' DataFrame contains actual values for the forecasted period)
forecast_prophet_test = forecast_prophet.tail(6)  # Get last 6 predictions for RMSE evaluation
rmse_prophet = np.sqrt(mean_squared_error(test['Sales'], forecast_prophet_test['yhat']))  # Ensure 'test' contains actual sales for comparison
print(f"Prophet RMSE: {rmse_prophet:.2f}")
```

#### <font color='magenta'>6. LSTM Model  </font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate sample data (monthly sales data for 3 years)
np.random.seed(0)
months = pd.date_range(start='2019-01-01', periods=36, freq='M')
sales = np.random.randint(100, 200, size=(36,)) + (np.linspace(0, 50, 36))  # trend + noise

# Create DataFrame
data = pd.DataFrame({'Date': months, 'Sales': sales})
data.set_index('Date', inplace=True)

# Split data into training and testing sets
train = data[:int(len(data) * 0.8)]
test = data[int(len(data) * 0.8):]

# Prepare data for LSTM (scaling and reshaping)
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train[['Sales']])

# Prepare X_train and y_train using a 6-month rolling window
X_train, y_train = [], []
for i in range(6, len(train_scaled)):
    X_train.append(train_scaled[i-6:i, 0])  # Use the last 6 observations as features
    y_train.append(train_scaled[i, 0])  # The target is the next months value

# Convert lists to arrays
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshape X_train to be 3D for LSTM input: (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

# Build the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model_lstm.fit(X_train, y_train, epochs=20, batch_size=32)

# Prepare test data (scale the test set using the same scaler)
test_scaled = scaler.transform(test[['Sales']])

# Prepare X_test using the same 6-month window
X_test = []
for i in range(6, len(test_scaled)):
    X_test.append(test_scaled[i-6:i, 0])

# Convert to numpy array
X_test = np.array(X_test)

# Reshape X_test to be 3D for LSTM input
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Make predictions with LSTM
forecast_lstm_scaled = model_lstm.predict(X_test)
forecast_lstm = scaler.inverse_transform(forecast_lstm_scaled)  # Inverse scale the predictions

# Plot predictions
plt.figure(figsize=(10, 6))
plt.plot(train.index, train['Sales'], label="Training Data")
plt.plot(test.index, test['Sales'], label="Test Data")
plt.plot(test.index[6:], forecast_lstm, label="LSTM Forecast", linestyle='--')  # Align forecast with test data
plt.title('LSTM Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Evaluate RMSE
rmse_lstm = np.sqrt(mean_squared_error(test['Sales'][6:].values, forecast_lstm))  # Only compare to the forecasted part
print(f"LSTM RMSE: {rmse_lstm:.2f}")
```
